# ☞ One-dimensional Data Sets 

The code snippet below sets the random seed for reproducibility, generates an array of 20 random numbers from a standard normal distribution, creates an array representing the indices of the elements in y, and plots the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the random seed for reproducibility
np.random.seed(1000)

# Generate an array of 100 random numbers from a standard normal distribution
y = np.random.standard_normal(100)

# Create an array representing the indices of the elements in y
x = np.arange(len(y))

# Plot the data
plt.plot(x, y)

# Display the plot
plt.show()
```

# ☞ Two-dimensional Data Sets 

Plotting two-dimensional data sets helps visualize relationships between variables, explore patterns, analyze correlations, compare data, evaluate models, and communicate insights effectively.

```python
# Generate a 2-dimensional array of shape (20, 2) with random numbers from a standard normal distribution
y = np.random.standard_normal((20, 2))

# Perform cumulative sum operation column-wise along the first axis
y = y.cumsum(axis=0)
```

The code above generates a 2-dimensional array of shape (20, 2) filled with random numbers from a standard normal distribution. Then, the cumsum(axis=0) function is applied along the first axis (axis=0), which performs the cumulative sum operation column-wise.

```python
import matplotlib.pyplot as plt

# Set the figure size
plt.figure(figsize=(15, 6))

# Plot the cumulative sum of the y array with green color and increased line width
plt.plot(y, color='green', linewidth=1.5)

# Set the x-axis label
plt.xlabel('index')

# Set the y-axis label
plt.ylabel('value')

# Set the plot title
plt.title('Plot with Two Data Sets')

# Display the plot
plt.show()
```

The modified plot provides a clear distinction between the two datasets using different colors, line widths, and marker styles. 

The inclusion of a legend adds clarity and context to the plot. This visually appealing representation helps viewers differentiate and interpret the datasets effectively.

```python
# Set the figure size
plt.figure(figsize=(15, 6))

# Plot the cumulative sum of the first column of the y array with cyan color and increased line width
plt.plot(y[:, 0], 'c', lw=1.5, label='1st')

# Plot the cumulative sum of the second column of the y array with blue color and increased line width
plt.plot(y[:, 1], 'b', lw=1.5, label='2nd')

# Plot the entire y array with red circles as markers
plt.plot(y, 'ro')

# Set the x-axis label
plt.xlabel('index')

# Set the legend to display the labels of the plotted lines
plt.legend(loc=0)

# Set the y-axis label
plt.ylabel('value')

# Set the plot title
plt.title('Plot with Two Data Sets')

# Display the plot
plt.show()
```

It's worth knowing that:

**loc**: Specifies the location of the legend in the plot. It can take various values such as:

- 'best': Automatically choose the optimal location for the legend.

- 'upper right', 'upper left', 'lower right', 'lower left': Place the legend in the respective corner of the plot.

- 'center', 'center left', 'center right', 'upper center', 'lower center': Place the legend along the respective edge or center of the plot.

```python
# Multiply the values in the first column of the y array by 100
y[:, 0] = y[:, 0] * 100
```

```python
# Set the figure size
plt.figure(figsize=(15, 6))

# Plot the cumulative sum of the first column of the y array with magenta color and increased line width
plt.plot(y[:, 0], 'c', lw=1.5, label='1st')

# Plot the cumulative sum of the second column of the y array with cyan color and increased line width
plt.plot(y[:, 1], 'b', lw=1.5, label='2nd')

# Plot the entire y array with red circles as markers
plt.plot(y, 'ro')


# Set the x-axis label
plt.xlabel('index')

# Set the legend to display the labels of the plotted lines
plt.legend(loc=0)

# Set the y-axis label
plt.ylabel('value')

# Set the plot title
plt.title('Plot with two differently scaled data sets')

# Display the plot
plt.show()
```

After multiplying the values in the first column of the y array by 100, the scaling of the y-axis changes. This scaling affects the visualization of the data sets in the plot. The first data set remains visually readable, while the second data set appears as a straight line, making it visually lost.

To address this issue and improve the visualization, we have two options:

* Using two y-axes: This approach involves assigning one data set to the left y-axis and the other data set to the right y-axis. It allows us to maintain separate scaling for each data set, making them visually distinguishable.

* Using two subplots: This approach involves creating two separate subplots, each displaying one data set. By having separate subplots, we can provide individual scaling and visual clarity to each data set.

Both options offer different ways to overcome the issue of visually lost data due to scaling differences. Depending on the specific requirements and context of the data, either approach can be chosen to ensure effective visualization and interpretation of the data sets.

```python
# Set the figure size
fig, ax1 = plt.subplots(figsize=(15, 6))

# Plot the cumulative sum of the first column of the y array with cyan color and increased line width, and label it as '1st'
ax1.plot(y[:, 0], 'y', lw=1.5, label='1st')

# Plot the entire y array of the first column with red circles as markers
ax1.plot(y[:, 0], 'ro')

# Set the x-axis label
ax1.set_xlabel('index')

# Set the y-axis label for the first data set
ax1.set_ylabel('1st')

# Set the legend for the first data set to display the labels of the plotted lines
ax1.legend(loc='upper left')

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Plot the cumulative sum of the second column of the y array with blue color and increased line width, and label it as '2nd'
ax2.plot(y[:, 1], 'g', lw=1.5, label='2nd')

# Plot the entire y array of the second column with red circles as markers
ax2.plot(y[:, 1], 'ro')

# Set the y-axis label for the second data set
ax2.set_ylabel('2nd')

# Set the legend for the second data set to display the labels of the plotted lines
ax2.legend(loc='upper right')

# Set the plot title
plt.title('Plot with two differently scaled data sets using two y-axes')

#plt.grid(False)

# Display the plot
plt.show()
```

## ⇥ Scatter Plots

```python
# Generate a 2-dimensional array of random numbers from a standard normal distribution
y = np.random.standard_normal((1000, 2))

# Create a new figure with a specified size
plt.figure(figsize=(20, 10))

# Plot the data points from the first column against the second column
# 'ro' specifies red circles as markers for the points
plt.plot(y[:, 0], y[:, 1], 'ro')

# Set the label for the x-axis
plt.xlabel('1st')

# Set the label for the y-axis
plt.ylabel('2nd')

# Set the title of the plot
plt.title('Scatter Plot')

# Display the scatter plot
plt.show()
```

The latter code snippet generates a 2-dimensional array of random numbers from a standard normal distribution. It then creates a scatter plot using matplotlib to visualize the data. The scatter plot displays the points from the first column of the array on the x-axis and the points from the second column on the y-axis. The points are represented as red circles. The x-axis is labeled as '1st', the y-axis as '2nd', and the plot is titled 'Scatter Plot'. The plot is displayed on the screen using *plt.show()*.

Matplotlib also provides a specific function **plt.scatter()** to perform this task. See below the updated code.

```python
# Create a new figure with a size of 20x10
plt.figure(figsize=(20, 10))

# Plot the data points from the first column against the second column
# Use red circles as markers for the points
plt.scatter(y[:, 0], y[:, 1], marker='o')

# Set the label for the x-axis
plt.xlabel('1st')

# Set the label for the y-axis
plt.ylabel('2nd')

# Set the title of the plot
plt.title('Scatter Plot')

# Display the scatter plot
plt.show()
```

Now, let's take it to a higher dimension !

We want to generate a scatter plot with points represented by red circles. The positions of the points are determined by the first two columns of y, and their colors are based on random integers in the array c. The plot will include labels for the axes and a colorbar.

```python
# Create a new figure with a size of 20x10
plt.figure(figsize=(20, 10))

# Generate an array of random integers between 0 and 10 with the same length as y
c = np.random.randint(0, 10, len(y))

# Plot the data points from the first column against the second column
# Use red circles as markers for the points
# Use the 'coolwarm' colormap to assign colors based on the values in the array c
plt.scatter(y[:, 0], y[:, 1], c=c, cmap='coolwarm', marker='o')

# Add a colorbar to indicate the mapping of colors to the values in the array c
plt.colorbar()

# Set the label for the x-axis
plt.xlabel('1st')

# Set the label for the y-axis
plt.ylabel('2nd')

# Set the title of the plot
plt.title('Scatter Plot')

# Display the scatter plot
plt.show()
```

## ⇥ Histograms

In the context of financial analysis, **scatter plots** are valuable for comparing the returns of two financial time series. By plotting the returns of one series on the x-axis and the returns of another series on the y-axis, scatter plots allow us to visually analyze the relationship and potential correlations between the two variables. This can help identify patterns, trends, or clusters that may exist in the data, enabling us to derive insights and make informed decisions.

In addition to scatter plots, **histograms** are another powerful visualization tool in finance. Histograms provide a graphical representation of the distribution of a single variable or dataset. In the context of financial analysis, histograms are commonly used to analyze the distribution of returns, prices, volumes, or other financial metrics. They allow us to understand the frequency and concentration of observations within different ranges or bins, providing insights into the shape, central tendency, and spread of the data. 

Histograms can reveal characteristics such as skewness, kurtosis, and the presence of outliers, aiding in risk assessment, asset valuation, and strategy development.

Let's make this clarification before proceeding.

**Histograms** are used to visualize the distribution of continuous variables, while **Bar plots** are used to compare discrete categories or groups. Histograms display the frequency or count of data points within specific ranges, while bar plots compare values associated with distinct categories.

```python
# Create a new figure with a size of 20x10
plt.figure(figsize=(30, 15))

# Plot a histogram of the data points from the first and second columns
# The label parameter assigns labels to the histogram bars for each column
# The bins parameter specifies the number of bins or intervals in the histogram
plt.hist(y, label=['1st', '2nd'], bins=25)

# Display a legend indicating the labels for each column
plt.legend(loc=0)

# Set the label for the x-axis
plt.xlabel('Value')

# Set the label for the y-axis
plt.ylabel('Frequency')

# Set the title of the plot
plt.title('Histogram')

# Display the histogram plot
plt.show()
```

## ⇥ The Integral Value

As a final illustraion of this section, let's look at a mathematical inspired plot that can be handled by the **matplotlib** library as well. 

The **Integral** can be interpreted as an area under a curve when integrating a non-negative function over an interval. 

It represents the exact area bounded by the function and the x-axis within that interval. The integral accounts for the area above the x-axis as positive and the area below the x-axis as negative, if applicable. 

The magnitude of the integral represents the absolute value of the area, while the sign indicates the orientation of the area.

To illustrate that, we graph the following integral:

$$\int_a^b f(x) \, dx,$$ 

with $f(x) = \dfrac{1}{2}e^x + 1$, $a = \dfrac{1}{2}$ and $b = \dfrac{3}{2}$.

We will do that in two steps:

First, we code the function $f(x)$.

```python
# Define the function f(x)
def func(x):
    return 0.5 * np.exp(x) + 1

# Set the limits of integration
a, b = 0.5, 1.5

# Generate x-values for plotting
x = np.linspace(0, 2)

# Calculate y-values using the function
y = func(x)

# Generate x-values within the limits of integration
Ix = np.linspace(a, b)

# Calculate y-values corresponding to the limits of integration
Iy = func(Ix)

# Create a list of vertices for the polygon representing the area under the curve
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
```

Second, the plotting itself. Let's go! I can't wait to get the result :)

```python
from matplotlib.patches import Polygon

fig, ax = plt.subplots(figsize=(20, 10))

# Plot the function curve
plt.plot(x, y, 'red', linewidth=2)

# Set the lower limit of the y-axis to 0
plt.ylim(bottom=0)

# Create a polygon patch representing the area under the curve
poly = Polygon(verts, facecolor='chartreuse', edgecolor='black')

# Add the polygon patch to the plot
ax.add_patch(poly)

# Add the integral notation as a text label at the center of the area
plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x) \mathrm{d}x$',
         horizontalalignment='center', fontsize=20)

# Add labels for the x-axis and y-axis
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

# Set the x-axis ticks and labels
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))

# Set the y-axis ticks and labels
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))

# Set the title of the plot
ax.set_title('Exponential function, Integral value, and LaTeX labels')

# Display the plot
plt.show()
```
